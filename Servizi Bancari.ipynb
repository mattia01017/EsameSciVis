```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
dataSet = pd.read_csv('Attività svolte su internet.csv')
dataSet = dataSet[dataSet['Classe di età']!='6 anni e più']
dataSet = dataSet[dataSet['Tipo dato']=='usare servizi bancari']
dataSet = dataSet[dataSet['Sesso']=='totale']
dataSet
```

```python
labels = ['6-10 anni', '11-14 anni', '15-17 anni', '18-19 anni', '20-24 anni', '25-34 anni', '35-44 anni', '45-54 anni', '55-59 anni', '60-64 anni', '65-74 anni', '75 anni e più']
years = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]

fig, axs = plt.subplots(len(labels)//4, 4)
fig.set_figheight(15)
fig.set_figwidth(25)

for i, l in enumerate(labels):
        ax = axs[i//4, i%4]
        ax.set_ylim(0, 100)
        ax.bar(
                dataSet.loc[dataSet['Classe di età'] == l, 'TIME'],
                dataSet.loc[dataSet['Classe di età'] == l, 'Value'],
                label = l
        )
        ax.set_xticks([years[i] for i in range(0, len(years), 2)])
        ax.set_title(l.capitalize())

plt.show()
```

```python
classe = ['6-10 anni']
data = dataSet.loc[list(map(lambda x: x in classe, dataSet['Classe di età'])) & (dataSet['Tipo dato'] == 'usare servizi bancari') & (dataSet['Sesso'] == 'totale')]

anno = data['TIME'].values
valori = data['Value'].values

fig, ax = plt.subplots()
ax.set_ylim(0,75)
ax.bar(anno, valori)

ax.set(xlabel = 'Anno', ylabel='Valori', title='6-10 anni')

plt.show()
```
