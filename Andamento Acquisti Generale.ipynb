```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.constants import golden
```

```python
acquisti = pd.read_csv('Acquisti o Ordini.csv', sep=',')
```

```python
pd.set_option('display.max_rows', acquisti.shape[0]+1)
dataframe = acquisti[['Tipo dato', 'Value', 'TIME', 'Classe di età', "Sesso"]]
```

```python
dataframe = dataframe[((dataframe['Classe di età']=="14 anni e più")&((acquisti['Tipo dato'] == "negli ultimi 3 mesi") | (acquisti['Tipo dato'] == "da 3 mesi a 1 anno fa"))&(acquisti['Sesso'] == "totale"))]
dataframe
```

```python
risultato = dataframe[["Tipo dato", "TIME", "Value"]]
risultato
```

```python
risultato = risultato.groupby(['TIME']).sum()
risultato
```

```python
plt.rc('figure', figsize=(11.0, 6.0))
plt.style.use('fivethirtyeight')
risultato.plot.bar(color='tab:blue', legend=False, alpha=0.6)
plt.title("Andamento Acquisti Online negli Anni")
plt.xlabel('')
plt.show()
```

```python
anni = risultato.index.values.tolist()
```

```python
valori = risultato["Value"].values.tolist()
```

```python
x = np.arange(len(anni))
fig, ax = plt.subplots()
ax.stackplot(anni, valori, labels=anni, alpha=1, color='Blue')
ax.set_title('Andamento Acquisti Online negli Anni')
ax.set_ylabel("Percentuale d'Acquisto Annuo")
```

```python
x = anni
y = valori

plt.stem(x, y)
plt.ylabel('Percentuale acquisto annuo', fontsize=16)
plt.show()
```
