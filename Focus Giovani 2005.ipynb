```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.constants import golden
```

```python
acquisti = pd.read_csv('Acquisti o Ordini.csv', sep=',')
```

```python
dataframe = acquisti[((acquisti['Tipo dato'] == "negli ultimi 3 mesi") | (acquisti['Tipo dato'] == "da 3 mesi a 1 anno fa")) & (acquisti['Sesso'] == "totale") & (acquisti['Classe di et√†'] != "14 anni e pi√π")]
dataframe = dataframe.drop(columns=["Sesso","TIPO_DATO_SDS_ICT", "SEXISTAT1", "ETA1", "Flag Codes", "TIME","Flags"], inplace=False)
pd.set_option('display.max_rows', dataframe.shape[0]+1)
dataframe
```

```python
dataframe_primafascia = dataframe[((dataframe['Classe di et√†'] == "14-17 anni") | (dataframe['Classe di et√†'] == "18-19 anni") | (dataframe['Classe di et√†'] == "20-24 anni"))]
dataframe_primafascia = dataframe_primafascia[((dataframe_primafascia['Seleziona periodo'] == 2020) | (dataframe_primafascia['Seleziona periodo'] == 2005))]
dataframe_primafascia = dataframe_primafascia.groupby(["Seleziona periodo"]).mean()
dataframe_primafascia
```

```python
dataframe_secondafascia = dataframe[((dataframe['Classe di et√†'] == "25-34 anni") | (dataframe['Classe di et√†'] == "35-44 anni")|(dataframe['Classe di et√†'] == "45-54 anni")|(dataframe['Classe di et√†'] == "55-59 anni"))]
dataframe_secondafascia = dataframe_secondafascia[((dataframe_secondafascia['Seleziona periodo'] == 2020) | (dataframe_secondafascia['Seleziona periodo'] == 2005))]
dataframe_secondafascia = dataframe_secondafascia.groupby(["Seleziona periodo"]).mean()
dataframe_secondafascia
```

```python
dataframe_terzafascia = dataframe[((dataframe['Classe di et√†'] == "60-64 anni") | (dataframe['Classe di et√†'] == "65-74 anni") | (dataframe['Classe di et√†'] == "75 anni e pi√π"))]
dataframe_terzafascia = dataframe_terzafascia[((dataframe_terzafascia['Seleziona periodo'] == 2020) | (dataframe_terzafascia['Seleziona periodo'] == 2005))]
dataframe_terzafascia = dataframe_terzafascia.groupby(["Seleziona periodo"]).mean()
dataframe_terzafascia
```

```python
#Focus Prima Fascia
dataframe_focus = dataframe[((dataframe['Classe di et√†'] == "14-17 anni") | (dataframe['Classe di et√†'] == "18-19 anni") | (dataframe['Classe di et√†'] == "20-24 anni"))]
dataframe_focus = dataframe_focus[((dataframe_focus['Seleziona periodo'] == 2020) | (dataframe_focus['Seleziona periodo'] == 2005))]
dataframe_focus = dataframe_focus.groupby(["Classe di et√†", "Seleziona periodo"]).sum()
dataframe_focus
```

```python
list = [dataframe_primafascia.at[2005, 'Value']]
list.append(dataframe_secondafascia.at[2005, 'Value'])
list.append(dataframe_terzafascia.at[2005, 'Value'])
list
```

```python
#Lista Focus 2005
list_focus = [dataframe_focus.at[("14-17 anni", 2005), "Value"]]
list_focus.append(dataframe_focus.at[("18-19 anni", 2005), "Value"])
list_focus.append(dataframe_focus.at[("20-24 anni", 2005), "Value"])
list_focus
```

```python
somma = sum(list)
somma
```

```python
somma_focus = sum(list_focus)
somma_focus
```

```python
from matplotlib.patches import ConnectionPatch

# make figure and assign axis objects
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 5))
fig.subplots_adjust(wspace=0)


# pie chart parameters
overall_ratios = [y*100/somma for y in list]
labels = ['Fascia 14-24', 'Fascia 25-59', 'Fascia 60 e oltre']
explode = [0.1, 0, 0]
# rotate so that first wedge is split by the x-axis
angle = -180 * overall_ratios[0]
wedges, *_ = ax1.pie(overall_ratios, autopct='%1.1f%%', startangle=angle,
                     labels=labels, explode=explode)

# bar chart parameters
age_ratios = [(y/somma_focus) for y in list_focus]
age_labels = ['14-17 anni', '18-19 anni', '20-24 anni']
bottom = 1
width = .2

# Adding from the top matches the legend.
for j, (height, label) in enumerate(reversed([*zip(age_ratios, age_labels)])):
    bottom -= height
    bc = ax2.bar(0, height, width, bottom=bottom, color='C0', label=label,
                 alpha=0.1 + 0.25 * j)
    ax2.bar_label(bc, labels=[f"{height:.0%}"], label_type='center')

ax2.set_title('Fascia Giovani Focus')
ax2.legend()
ax2.axis('off')
ax2.set_xlim(- 2.5 * width, 2.5 * width)
plt.show()
```
