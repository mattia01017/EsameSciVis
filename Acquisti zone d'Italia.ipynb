```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.constants import golden

plt.rc('figure', figsize=(10.0, 5.0))
```

```python
acquisti = pd.read_csv('Acquisti o Ordini - reg. e tipo di comune.csv', sep=',')
```

```python
pd.set_option('display.max_rows', acquisti.shape[0]+1)
acquisti = acquisti[["Territorio", "Tipo dato", "TIME", "Value"]]
acquisti = acquisti[((acquisti['Tipo dato'] == "negli ultimi 3 mesi") | (acquisti['Tipo dato'] == "da 3 mesi a 1 anno fa"))]
acquisti = acquisti[((acquisti["Territorio"]=="Nord-est") | (acquisti["Territorio"]=="Nord-ovest") | (acquisti["Territorio"]=="Centro") | (acquisti["Territorio"]=="Sud") | (acquisti["Territorio"]=="Isole"))]
acquisti
```

```python
#Centro
centro = acquisti[acquisti["Territorio"]=="Centro"]
centro = centro.groupby(['TIME']).sum()
centro
```

```python
#Valori Centro
valori_centro = centro["Value"].values.tolist()
valori_centro
```

```python
#Sud
sud = acquisti[acquisti["Territorio"]=="Sud"]
sud = sud.groupby(['TIME']).sum()
sud
```

```python
#Valori Sud
valori_sud = sud["Value"].values.tolist()
valori_sud
```

```python
#Nord
nord = acquisti[((acquisti["Territorio"]=="Nord-est") | (acquisti["Territorio"]=="Nord-ovest"))]
nord = nord.groupby(['TIME', "Territorio"]).sum()
nord = nord.groupby(["TIME"]).mean()
nord
```

```python
#Valori Nord
valori_nord = nord["Value"].values.tolist()
valori_nord
```

```python
#Isole
isole = acquisti[(acquisti["Territorio"]=="Isole")]
isole = isole.groupby(['TIME', "Territorio"]).sum()
isole = isole.groupby(["TIME"]).mean()
isole
```

```python
#Valori Isole
valori_isole = isole["Value"].values.tolist()
valori_isole
```

```python
#Anni
anni = acquisti["TIME"].values.tolist()
anni = anni[:16]
anni
```

```python
#Nord
width = 0.35      # the width of the bars: can also be len(x) sequence

labels = anni
Nord = valori_nord
fig, ax = plt.subplots()

ax.bar(labels, Nord, width, label='Nord', alpha = 0.8, color="brown")

ax.set_ylabel('Percentuale acquisto Annuo')
ax.set_title('Nord Italia')
plt.rc('figure', figsize=(10.0, 5.0))
plt.show()
```

```python
#Centro
labels = anni
Centro = valori_centro
fig, ax = plt.subplots()

ax.bar(labels, Centro, width, label='Centro', alpha = 1, color="olive")

ax.set_ylabel('Percentuale Acquisto Annuo')
ax.set_title('Centro Italia')

plt.rc('figure', figsize=(10.0, 5.0))
plt.show()
```

```python
#Sud
labels = anni
Sud = valori_sud
fig, ax = plt.subplots()

ax.bar(labels, Sud, width, label='Sud', alpha = 0.8, color="Orange")

ax.set_ylabel('Percentuale Acquisto Annua')
ax.set_title('Sud Italia')
plt.rc('figure', figsize=(10.0, 5.0))
plt.show()
```

```python
#Isole
labels = anni
Isole = valori_isole
fig, ax = plt.subplots()

ax.bar(labels, Isole, width, label='Isole', alpha = 1, color="darkturquoise")

ax.set_ylabel('Percentuale Acquisto Annua')
ax.set_title("Isole d'Italia")
plt.rc('figure', figsize=(10.0, 5.0))
plt.show()
```

```python
labels = anni
Centro = valori_centro
Sud = valori_sud
Nord = valori_nord
Isole = valori_isole

x = np.arange(len(labels))  # the label locations

f, ax = plt.subplots()

ax.set_ylabel("Percentuale Acquisto Annuo")
ax.plot(x, Nord, label='Nord', marker='o', color="brown")
ax.plot(x, Centro, label='Centro', marker='o', color="olive")
ax.plot(x, Sud, label='Sud', marker='o', color="Orange")
ax.plot(x, Isole, label='Isole', marker='o', color="darkturquoise")

ax.set_xticks(x, labels)
ax.legend()
ax.grid(visible=True, linestyle="--")
plt.show()
```
