```python
viaggi = dataSet[dataSet['Tipo dato']=='usare servizi relativi a viaggi o soggiorni']
viaggi = viaggi[viaggi['TIME']!=2020]
```

```python
viaggi = viaggi[viaggi['Sesso']=='totale']
viaggi610 = viaggi[viaggi['Classe di età']=='6-10 anni']
viaggi1114 = viaggi[viaggi['Classe di età']=='11-14 anni']
viaggi1517 = viaggi[viaggi['Classe di età']=='15-17 anni']
viaggi1819 = viaggi[viaggi['Classe di età']=='18-19 anni']
viaggi2024 = viaggi[viaggi['Classe di età']=='20-24 anni']
viaggi2534 = viaggi[viaggi['Classe di età']=='25-34 anni']
viaggi3544 = viaggi[viaggi['Classe di età']=='35-44 anni']
viaggi4554 = viaggi[viaggi['Classe di età']=='45-54 anni']
viaggi5559 = viaggi[viaggi['Classe di età']=='55-59 anni']
viaggi6064 = viaggi[viaggi['Classe di età']=='60-64 anni']
viaggi6574 = viaggi[viaggi['Classe di età']=='65-74 anni']
viaggi75 = viaggi[viaggi['Classe di età']=='75 anni e più']
```

```python
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2020]
population_by_continent = {
    '6-10 anni': viaggi610['Value'].values,
    '11-14 anni': viaggi1114['Value'].values,
    '15-17 anni': viaggi1517['Value'].values,
    '18-19 anni': viaggi1819['Value'].values,
    '20-24 anni': viaggi2024['Value'].values,
    '25-34 anni': viaggi2534['Value'].values,
    '35-44 anni': viaggi3544['Value'].values,
    '45-54 anni': viaggi4554['Value'].values,
    '55-59 anni': viaggi5559['Value'].values,
    '60-64 anni': viaggi6064['Value'].values,
    '65-74 anni': viaggi6574['Value'].values,
    '75 anni e più': viaggi75['Value'].values,
}

fig, ax = plt.subplots()
ax.stackplot(year, population_by_continent.values(),
             labels=population_by_continent.keys(), alpha=1)
plt.legend()
sns.move_legend(ax, "lower center", bbox_to_anchor=(.5, 1.05), ncol=6, title=None, frameon=False)
ax.set_title('Usare servizi relativi a viaggi o soggiorni ')
ax.set_xlabel('Year')
ax.set_ylabel('Valori')

plt.show()
```
