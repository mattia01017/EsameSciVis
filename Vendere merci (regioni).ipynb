```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
data = pd.read_csv('Attività_regioni.csv')
zones = ['Nord-ovest', 'Nord-est', 'Centro', 'Sud', 'Isole']
years=data['TIME'].unique()

data = data.loc[list(map(lambda x: x in zones, data['Territorio']))]
data = data[data['Tipo dato']=='vendere merci o servizi (es. aste online, eBay)']
data
```

```python
population = {
    'Nord-ovest': data[data['Territorio']=='Nord-ovest']['Value'].values,
    'Nord-est': data[data['Territorio']=='Nord-est']['Value'].values,
    'Centro': data[data['Territorio']=='Centro']['Value'].values,
    'Sud': data[data['Territorio']=='Sud']['Value'].values,
    'Isole': data[data['Territorio']=='Isole']['Value'].values,
}

fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(8)
ax.stackplot(years, population.values(),
             labels=population.keys(), alpha=0.8)
ax.legend(loc='upper left')
ax.set_xlabel('Anni')
ax.set_ylabel('% utenti che vendono merci o servizi')

plt.show()
```
