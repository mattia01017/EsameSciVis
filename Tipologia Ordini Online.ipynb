```python
%matplotlib inline

import csv
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.constants import golden
```

```python
ordini = pd.read_csv('Tipo di Acquisti e Ordini.csv', sep=',')
```

```python
pd.set_option('display.max_rows', ordini.shape[0]+1)
ordini = ordini[["Tipo dato", "Sesso", "Classe di et√†", "TIME", "Value"]]
ordini
```

```python
#Tipologie di Acquisti

df_tipologieacquisti = ordini[((ordini["Sesso"]=='totale') & (ordini["Classe di et√†"]=="14 anni e pi√π"))]
df_tipologieacquisti = df_tipologieacquisti.groupby("Tipo dato").mean()
df_tipologieacquisti = df_tipologieacquisti.drop(columns=["TIME"])
df_tipologieacquisti = df_tipologieacquisti.drop(["altro", "altre spese di viaggio per vacanze"])
df_tipologieacquisti.sort_values(by=['Value'], ascending=False, inplace=True)
df_tipologieacquisti
```

```python
Valori_Medi = df_tipologieacquisti["Value"].values.tolist()
```

```python
categorie = df_tipologieacquisti.index.values.tolist()
categorie
```

```python
#Tipologie di Acquisti
labels = categorie

plt.rc('figure', figsize=(10.0, 5.0))
fig, ax = plt.subplots()
width = 0.35 
plt.grid()
ax.bar(labels, Valori_Medi, width, label='Tipologie di Acquisti', alpha = 0.6, color = 'forestgreen')

plt.xticks(rotation='vertical')
ax.set_ylabel('% Persone')
ax.set_title('Tipologie di Acquisti')
plt.show()
```

```python
plt.rc('figure', figsize=(11.0, 10.0))
fig, ax = plt.subplots()
y_pos = np.arange(len(categorie))
performance = Valori_Medi
width = 0.40 
ax.barh(y_pos, performance, width, align='center', color = 'forestgreen', alpha=0.6)
ax.set_yticks(y_pos, labels=categorie)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Percentuale Acquisto annuo')
ax.set_title('Tipologie di Prodotti')
plt.grid()
plt.show()
```
